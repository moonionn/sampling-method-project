```python
import numpy as np
import matplotlib.pyplot as plt

def plot_model_performance(sampling, f1_scores, recalls, g_means, mccs, aucs):

    # 将数据组合成一个数组，方便后续处理
    data = np.array([f1_scores, recalls, g_means, mccs, aucs])

    # 指标名称
    metrics = ['F1-score', 'Recall', 'G-mean', 'MCC', 'AUC']

    # 创建条形图
    fig, ax = plt.subplots()

    # 设置条形图的位置和宽度
    bar_width = 0.15
    index = np.arange(len(sampling))

    # 为每个指标创建一个条形图
    for i in range(len(metrics)):
        ax.bar(index + i * bar_width, data[i], bar_width, label=metrics[i])

    # 添加图表标签和标题
    ax.set_xlabel('Models')
    ax.set_ylabel('Scores')
    ax.set_title('Performance Metrics for Different Models')
    ax.set_xticks(index + bar_width * (len(metrics)-1) / 2)
    ax.set_xticklabels(sampling)

    # 添加图例
    ax.legend(loc='lower right')

    # 展示图表
    plt.show()

# Example usage:
sampling = ['SMOTE', 'ADASYN', 'ROS', 'RUS']
f1_scores = [0.905, 0.855, 0.904, 0.905]
recalls = [0.877, 0.942, 0.878, 0.876]
g_means = [0.883, 0.872, 0.883, 0.883]
mccs = [0.562, 0.479, 0.562, 0.562]
aucs = [0.962, 0.962, 0.962, 0.961]

plot_model_performance(sampling, f1_scores, recalls, g_means, mccs, aucs)
```
